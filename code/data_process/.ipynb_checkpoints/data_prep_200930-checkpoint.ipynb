## Compile Sequence from Cymobase

```python
pattern = '../data/cymobase/seq/*.fasta'
files = glob.glob(pattern)
```

```python
cymobase = pd.DataFrame()
```

```python
for file in files[0:10]:
    print(file)
    records = SeqIO.parse(file, "fasta")
    label = re.split(r"[_.]",file)[-2]
    ids = []
    descriptions = []
    seqs = []
    
    for record in records:
        ids.append(record.id)
        descriptions.append(record.description)
        seqs.append(record.seq)
        tmp_df = pd.DataFrame({'id': ids,'description': descriptions, 'seq': seqs})
        tmp_df["seq"] = tmp_df["seq"].apply(lambda s: s.__str__())
        tmp_df["type"] = [label]*len(ids)
        cymobase = pd.concat([cymobase,tmp_df],ignore_index=True)
```

```python
for file in files[10:20]:
    print(file)
    records = SeqIO.parse(file, "fasta")
    label = re.split(r"[_.]",file)[-2]
    ids = []
    descriptions = []
    seqs = []
    
    for record in records:
        ids.append(record.id)
        descriptions.append(record.description)
        seqs.append(record.seq)
        tmp_df = pd.DataFrame({'id': ids,'description': descriptions, 'seq': seqs})
        tmp_df["seq"] = tmp_df["seq"].apply(lambda s: s.__str__())
        tmp_df["type"] = [label]*len(ids)
        cymobase = pd.concat([cymobase,tmp_df],ignore_index=True)
```
