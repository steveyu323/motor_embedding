```python
from Bio import SeqIO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm 
import glob
import re
import requests
import io
```

```python
pfamA_motors = pd.read_csv("../../data/pfamA_motors.csv")
pfamA_motors.head()
```

```python
pfamA_motors["uniprot_acc"] = pfamA_motors["id"].apply(lambda s: s.split("/")[0])
pfamA_motors.head()
```

```python
def get_uniprot_entry(acc):
    url = "https://www.uniprot.org/uniprot/?query="+acc+"&sort=score&columns=id,entry name,reviewed,protein names,genes,organism,organism-id,lineage(all),sequence,length&format=tab"
    try:
        urlData = requests.get(url).content
        rawData = pd.read_csv(io.StringIO(urlData.decode('utf-8')),sep = '\t')
    except:
        rawData = pd.DataFrame()
    return rawData 
```

```python
motor_uniprot = pd.read_csv("../../data/motor_uniprot.csv")
```

```python
motor_uniprot = get_uniprot_entry(pfamA_motors["uniprot_acc"].iloc[0])
for i in range(1,pfamA_motors.shape[0]):
    curr = get_uniprot_entry(pfamA_motors["uniprot_acc"].iloc[i])
    if not (curr.empty):
        motor_uniprot = pd.concat([motor_uniprot,curr], ignore_index=True)
    if i%1000 ==0:
        print(motor_uniprot.shape)
        motor_uniprot.to_csv("../../data/motor_uniprot.csv",index = False)
```
