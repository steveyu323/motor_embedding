3. Background on motor and microtubule
    1) microtubules are largest components of the cytoskeleton, and are polymers of tubulin dimers
    2) centrosome is the main microtubule organizing center of animal cells composed of two centrioles
    3) Motor proteins hydrolyze ATP and cause conformational change that produce movement. Example: kinesin transports vesicles along microtubule tracks

    1. UniRep is a unidirectional mLSTM with 1900 units
